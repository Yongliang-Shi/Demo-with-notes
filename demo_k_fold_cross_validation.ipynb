**Intro**
- In machine learning, we usually divide the dataset into train, validate and test.
    - training: 60% 
    - validation: 10 - 20%
    - test: 5 - 20% 
    
<font color=red>**What happened if there are only training and test?**</font>
- fail to generalize on the test due to sample variablity

**What is cross-validation?**
- a statistical techinique which partitioning data into subests, training the data on a subset and use the other subset to evalute the model's performance. 

**K fold cross validate**
- randomly dividing the dataset into k groups or folds of about equal size. 
- the value o of K is typicllaly 5 or 10. 

**Stratified cross-validation**
- each fold has a good representation of the whole dataset.
- force each fold to have at least m instances of each class.
- ensure that one class of data it not overrepresented especially when the target variable if unbalanced. 

```python
# Import libraries

import numpy as np
import pandas as pd
from sklearn.model_selection import KFold
from sklearn.preprocessing import MinMaxScaler
from sklearn.svm import SVR

import warnings
warnings.filterwarnings("ignore")
```

```python
# Read the Boston House price dataset
# sep='\s+' defines separator as being one single white space or more. 

url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/housing.data'
df = pd.read_csv(url, sep='\s+', header=None)
df.head()
```

```python
# Check the shape of the dataset
df.shape
```

```python
# Specify the features and the output variable of the dataset

X = df.iloc[:, 0:13]
y = df.iloc[:, 13]

# Check the shape of X and y
X.shape, y.shape
```

```python
# Check the top 5 rows of X
X.head()
```

```python
# Check the top 5 values of y
y.head()
```

```python
# Scale the feature values between 0 and 1
# Create the scaler object

scaler = MinMaxScaler()
X = scaler.fit_transform(X)
# X now is a np.array
```

```python
# K-Fold CV

scores=[]

# Using the RBF kernel of the SVR model with default parameter values
best_svr = SVR(kernel='rbf')

# Set the number of folds to 10
cv = KFold(n_splits=10, random_state=42)

for train_index, test_index in cv.split(X):
    
    # Print out the indexes of the training and the testing sets for each iteration
    print("Train Index:", train_index, "\n")
    print("Test Index:", test_index)
    
    # Specify train and test datasets for each iteratoin
    X_train = X[train_index]
    y_train = y[train_index]
    X_test = X[test_index]
    y_test = y[test_index]
    
    # Fit the model to the training data
    best_svr.fit(X_train, y_train)
    
    # Return the coefficient of determination r2 of the prediction
    score = best_svr.score(X_test, y_test)
    
    # Append the error metric r2 to a list
    scores.append(score)
```

```python
# Print out the list of scores
scores
```

```python
# Compute the mean of r2 scores to determine the overall accuracy of the model
print(np.mean(scores))
```

```python
# Another way to give a list of r2 scores

from sklearn.model_selection import cross_val_score
cross_val_score(best_svr, X, y, cv=10)
```

```python
# Give a list of predictions

from sklearn.model_selection import cross_val_predict
cross_val_predict(best_svr, X, y, cv=10)
```
