```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# x is the probability between 0 and 1
# y = x/(1-x)

x = np.random.uniform(low=0.0, high=1.0, size=100)
y = x/(1-x)

# What if log transform y?
log_y = np.log10(y)

# Scatterplot y and log_y against x
plt.figure(figsize=(13,7))

plt.subplot(121)
plt.scatter(x, y)
plt.axhline(0, color='k', linestyle='dashed', linewidth=1)

plt.subplot(122)
plt.scatter(x, log_y)
plt.axhline(0, color='k', linestyle='dashed', linewidth=1)
```

```python
# Plot the distribution of y and log_y

plt.figure(figsize=(13,7))

plt.subplot(121)
pd.Series(y).hist()

plt.subplot(122)
pd.Series(log_y).hist()
```
